#### Load libraries and set random number generator seed

```python
import numpy as np
import pandas as pd
from sklearn import datasets
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Perceptron
from sklearn.metrics import plot_confusion_matrix
from sklearn.datasets import load_digits
from sklearn.metrics import ConfusionMatrixDisplay
```

```python
np.random.seed(42)
```

#### Code the solution

```python
# 1. Load the data. 

X, Y = load_digits(return_X_y=True, as_frame=True)
X
```

```python
X = X.astype('float32') / 255
X
```

```python
Y
```

```python
Y_onehot = pd.get_dummies(Y)
Y_onehot
```

```python
for digit in range(10):
  X_train, X_test, Y_train, Y_test = train_test_split(X, Y_onehot[digit], test_size=0.2, random_state=0, stratify=Y_onehot[digit])
  model = Perceptron().fit(X_train, Y_train)
  score = model.score(X_test, Y_test)

  # non-normalized confusion matrix
  ConfusionMatrixDisplay.from_estimator(model, X_test, Y_test,
                                        cmap=plt.cm.Blues,
                                        normalize=None).ax_.set_title(f'Test set results for {digit}: ({score:0.2})')
```
